```python
import turtle
a = 100
b = 200
d = 50
br = 'brown'
g1 = 'dark green'
g2 = 'green'
g3 = 'light green'
t = turtle.Turtle()
t.speed(10)
t.penup()
t.goto(d,-100)
t.pendown()
t.pensize(3)
t.pencolor('light blue')
t.bk(b)
t.lt(90)
t.fd(a)
t.rt(90)
t.fd(b)
t.rt(90)
t.fd(a)
t.penup()
t.goto(-140,0)
t.pendown()
t.lt(135)
t.fd(a)
t.rt(90)
t.fd(a)
t.penup()
t.goto(-34,35)
t.pendown()
t.rt(45)
t.bk(d)
t.lt(90)
t.fd(20)
t.rt(90)
t.fd(70)
t.penup()
t.goto(-30,120)
t.pendown()
t.fd(30)
t.penup()
t.goto(-22,135)
t.pendown()
t.fd(45)
t.penup()
t.goto(-75,-100)
t.pendown()
t.bk(60)
t.lt(90)
t.fd(40)
t.rt(90)
t.fd(60)

t.penup()
t.goto(-250,-150)
t.color(br)
t.begin_fill()
t.bk(50)
t.lt(90)
t.fd(20)
t.rt(90)
t.fd(50)
t.rt(90)
t.fd(20)
t.end_fill()
t.goto(-310,-100)
t.color(g1)
t.begin_fill()
t.bk(a+25)
t.rt(30)
t.fd(65)
t.end_fill()
t.goto(-295,-70)
t.color(g2)
t.begin_fill()
t.lt(30)
t.bk(100)
t.rt(30)
t.fd(55)
t.lt(60)
t.end_fill()
t.color(g3)
t.goto(-280,-45)
t.begin_fill()
t.rt(30)
t.bk(70)
t.rt(40)
t.fd(40)
t.end_fill()

t.goto(240,-100)
t.color(br)
t.begin_fill()
t.lt(40)
t.bk(30)
t.rt(90)
t.fd(a)
t.lt(90)
t.fd(30)
t.end_fill()
t.goto(150,0)
t.color(g1)
t.begin_fill()
t.bk(200)
t.rt(30)
t.fd(110)
t.end_fill()

t.lt(30)
t.fd(75)
t.color(g2)
t.begin_fill()
t.bk(150)
t.rt(40)
t.fd(95)
t.end_fill()

t.lt(40)
t.fd(50)
t.color(g3)
t.begin_fill()
t.bk(a)
t.rt(50)
t.fd(70)
t.end_fill()

```
